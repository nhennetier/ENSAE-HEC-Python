### 1. Importer pandas

```python
import pandas as pd
```

### Voici les données sous la forme d'un dictionnaire

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

### 2. Créer un dataframe avec ces données

#### Si vous n'indiquez pas le nom des colonnes dans le bon ordre, pandas les trieras automatiquement par ordre alphabétique

```python
army = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'deaths', 'battles', 'size', 'veterans', 'readiness', 'armored', 'deserters', 'origin'])
```

### 3. Mettez la colonne `origin` en index

```python
army = army.set_index('origin')
army
```

### 4. Afficher uniquement la colonne `veterans`

```python
army['veterans']
```

### 5. Afficher les colonnes `veterans` et `deaths`

```python
army[['veterans', 'deaths']]
```

### 6. Selectionnez les colonnes `deaths`, `size` et `deserters` de `Maine` et `Alaska`

```python
# Select all rows with the index label "Maine" and "Alaska"
army.loc[['Maine','Alaska'] , ["deaths","size","deserters"]]
```

### 7. Selectionner les lignes 3 à 6 et les colonnes 3 à 5

```python
#
army.iloc[3:7, 3:6]
```

### 8. Selectionner toutes les lignes à partir de la 4eme.

```python
army.iloc[3:]
```

### 9. Selectionner toutes les lignes jusqu'à la 4eme.

```python
army.iloc[:3]
```

### 10. Selectionner les colonnes entre la 4eme et la 6eme.

```python
# the first : means all
# after the comma you select the range

army.iloc[: , 4:7]
```

### 11. Selectionner les lignes avec plus de 50 morts

```python
army[army['deaths'] > 50]
```

### 12. Selectionner les lignes avec plus de 500 morts ou moins de 50

```python
army[(army['deaths'] > 500) | (army['deaths'] < 50)]
```

### 13. Selectionner les régiments qui ne se nomment pas `Dragoons`

```python
army[(army['regiment'] != 'Dragoons')]
```

### 14. Sélectionner les lignes nommées `Arizona` et `Texas`

```python
army.loc[['Arizona', 'Texas']]
```

### 15. Sélectionner la 3eme cellule de la ligne `Arizona`

```python
army.loc['Arizona', army.columns[2]]
```

### 16. Sélectionner la 2eme cellule de la colonne deaths

```python
army.iloc[[2], army.columns.get_loc('deaths')]
```

### 17. Trier les lignes par `size` puis par `deaths`

```python
army.sort_values(['size', 'deaths'])
```
